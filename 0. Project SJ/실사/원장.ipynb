```python
import pandas as pd
import numpy as np
```

```python
l = pd.read_excel('./원장/Socket/계정별원장(2021).xlsx')

#자료
l_2018 = pd.read_excel('./원장/Socket/계정별원장(2018).xlsx', sheet_name = None)
l_2019 = pd.read_excel('./원장/Socket/계정별원장(2019).xlsx', sheet_name = None)
l_2020 = pd.read_excel('./원장/Socket/계정별원장(2020).xlsx', sheet_name = None)
l_2021 = pd.read_excel('./원장/Socket/계정별원장(2021).xlsx', sheet_name = None)

l
```

```python
# 기본설정
월계 = l.iloc[3032,1]
누계 = l.iloc[3033,1]
계정명행 = 1
계정명열 = 6
시작행 = 3

head = l.loc[시작행-1].tolist()

head_replaced = []
for i in head:
    head_replaced.append(i.replace(" ",""))
    
print(head_replaced)
print(월계)
print(누계)
```

```python
#데이터프레임생성
df = pd.DataFrame(columns = head_replaced)
df['년'] = ""
df['계정명'] = ""
```

### 2018년

```python
sheets = list(l_2018.keys())

#시트설정
for i in range(len(sheets)):
    sheet = l_2018[sheets[i]]
    
    year = 2018
    계정명 = sheet.iloc[계정명행,계정명열]
    계정명 = 계정명.replace("계정과목","")
    계정명 = 계정명.replace(":","")
    계정명 = 계정명.replace(" ","")
    
    #데이터프레임 추가
    for j in range(시작행,len(sheet)):
        if 월계 in sheet.loc[j].tolist() or 누계 in sheet.loc[j].tolist():
            pass
        else:
            row = sheet.loc[j].tolist()
            row.append(year)
            row.append(계정명)
            
            df.loc[len(df)] = row
```
