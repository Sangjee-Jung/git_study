### DataFrame 뿌수고 합치기

```python
import pandas as pd
```

```python
f_ledger = pd.read_excel('./ledger.xlsx')
```

```python
f_ledger
```

```python
columns = f_ledger.columns
columns

rows = []

for column in columns:
    rows.append(f_ledger[column].tolist())

rows
```

```python
columns[0]
```

```python
rows[0]
```

```python
df = pd.DataFrame()
df
```

```python
for i in range(len(columns)):
    df[columns[i]] = rows[i]

df
```
